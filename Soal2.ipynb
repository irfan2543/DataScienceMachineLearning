Untuk pengelompokkan data berdasarkan Kategori Norma Hipotetik yang digunakan.
dengan menggunakan Norma Hipotetik menggunakan rumus sebagai berikut :
Mean = Nilai Tertinggi Persentase + Nilai Terendah Persentase / 2 = 100 + 0 / 2 = 50
Standar Deviasi = Nilai Tertinggi Persentase + Nilai Terendah Persentase / 6 = 100 + 0 / 6 = 17 (Dibulatkan)

Jadi dari hasil diatas didapatkan rumus berikut :

Low = X < 50 - 17 = X < 33%
Medium = 50 - 17 >= X <= 50 + 17 = 33% >= X <= 67%
High = X > 50 + 17 = X > 67%

bisa dilihat pada excel data berjumlah 9994 dengan kita asumsikan bahwa persentase untuk perhitungan adalah 100%.
dari data tersebut dapat kita bagi menjadi 3 kategori dalam frekuensi distribusinya yaitu :
  1. Kategori dengan Segment High dengan Range persentase yaitu > 67%
  2. Kateogri dengan Segment Low dengan Range persentase yaitu < 33%
  3. Kategori dengan Segment Medium dengan Range persentase diantara Low dan High yaitu >=33% dan <=67%

```python
#Membuat range antara nilai persentase maksimum dan minimum untuk Sales
high_percentile = 0.67
low_percentile = 0.33
```

```python
#Memanggil nilai dari quantile yang akan di hitung untuk menentukan median dari dataframe
sales_quantile = df_super['Sales'].quantile([high_percentile, low_percentile])
```

```python
#Melihat hasil nilai tertinggi dan terendah untuk dataframe
sales_quantile
```

```python
'''Membuat Fungsi untuk melakukan pembagian antar segment dengan berdasarkan :
  dari nilai maksimum sales dengan data diatas 0.70 dan didapatkan nilai terkecil adalah 129.9200
  dari nilai minimum sales dengan diatas 0.30 dan didapatkan nilai terbesar adalah 24.8707
'''
def segment1(sales):

    if sales > sales_quantile[high_percentile]:
        return "Segment High"
    elif sales > sales_quantile[low_percentile]:
        return "Segment Medium"
    else:
        return "Segment Low"
```

```python
#Melakukan pemanggilan menggunakan apply lambda untuk memasukkan fungsi segment kedalam column baru yaitu column Segment Customer
df_super['Segment Customer'] = df_super['Sales'].apply(lambda x: segment1(x))
```

```python
#Melakukan Order Date hanya tahun 2017
df_super['Order Date'] = pd.to_datetime(df_super['Order Date'])
order_date = (df_super['Order Date'].dt.year == 2017)
```

```python
#Memanggil df_super pada order_date
df_super[order_date]
```

```python
#Melakukan Perhitungan untuk Jumlah Customer pada tahun 2017
hasil_jumlh_customer = df_super[order_date].groupby('Segment Customer')['Order Date'].apply(lambda x: (x=='2017').count()).reset_index(name='Jumlah 2017')
```
